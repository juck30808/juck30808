```python
import numpy as np
import cv2
img = cv2.imread("input/Sample.png")
img.shape
```

```python
# pop-up
cv2.imshow('input/Sample.png', img)  
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# matplotlib
from matplotlib import pyplot as plt
plt.imshow(img[:,:,::-1]) # opencvBGR
plt.show()
```

```python
# method1 .cv2
img_gray = cv2.cvtColor(img, cv2.IMREAD_GRAYSCALE)
```

```python
# method2 . RGB  (R*0.299 + G*0.587 + B*0.114)
img_gray2 = np.array((img[:,:,0]*0.299 + img[:,:,0]*0.587 + img[:,:,0]*0.114), dtype="uint8")
```

```python
# Gray scale
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Gray scale using cv2 vs. formula')
ax1.imshow(img_gray, cmap = 'gray')
ax2.imshow(img_gray2, cmap = 'gray')
plt.show()
```

```python
hsv=cv2.cvtColor(img, cv2.COLOR_BGR2HSV) # HSV
h,s,v = cv2.split(hsv) # 

fig, (ax1, ax2,ax3) = plt.subplots(1, 3)

fig.suptitle('HSV')
ax1.imshow(h)
ax1.set_title("H")
ax2.imshow(s)
ax2.set_title("S")
ax3.imshow(v,cmap="gray")
ax3.set_title("V")
plt.show()
```

```python
import numpy as np
x, y = np.meshgrid(np.linspace(-1,1,7), np.linspace(-1,1,7)) # 
d = np.sqrt(x*x+y*y)
sigma, mu = 1.0, 0.0
kernel = np.exp(-( (d-mu)**2 / ( 2.0 * sigma**2 ) ) ) # 
kernel /= sum(sum(kernel)) #1
plt.imshow(kernel)
plt.show()
```
